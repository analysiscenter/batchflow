```python
# Necessary imports
import sys
import numpy as np

sys.path.append('../..')
from batchflow import plot
```

```python
arr = np.random.rand(10, 10)
```

```python
data = arr
title = 't'
cmap = 'ocean'
plot(data=data, title=title, cmap=cmap)
```

```python
data = [arr]
title = ['t']
cmap = ['ocean']
plot(data=data, title=title, cmap=cmap)
```

```python
data = [arr, arr]
title = ['t', 'y']
cmap = ['ocean', 'viridis']
plot(data=data, title=title, cmap=cmap)
```

```python
data = [arr, arr]
title = ['t', 'y']
cmap = ['ocean', 'viridis']
plot(data=data, title=title, cmap=cmap, separate=True)
```

```python
data = [arr, [arr, arr]]
title = ['t', 'y']
cmap = ['ocean', 'viridis']
plot(data=data, title=title, cmap=cmap)
```

```python
data = [arr, [arr, arr]]
title = ['t', ['y', 'x']]
cmap = ['ocean', ['viridis', 'Reds']]
plot(data=data, title=title, cmap=cmap)
```

```python
data = [arr, [arr, arr > .5]]
title = ['t0', 't1']
cmap = ['ocean', ['viridis', 'firebrick']]
plot(data=data, title=title, cmap=cmap)
```
