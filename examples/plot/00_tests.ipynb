```python
# Necessary imports
import sys
import numpy as np

sys.path.append('../..')
from batchflow import plot
```

```python
arr = np.random.randint(0, 9, (10, 20))
```

```python
plot(data=[arr, arr],
     combine='separate',
     title=['t1', 't2'],
     cmap=['Greys', 'Blues']
    )
```

```python
plot(data=[arr, arr],
     combine='separate',
     title=['t1', 't2'],
     cmap=['Greys', 'Blues'],
     mask_values=[None, lambda x: x > 5],
     colorbar=True,
    )
```

```python
plot(data=[arr, arr],
     combine='overlay',
     title='t',
     cmap=['Greys', 'Blues'],
     mask_values=[None, lambda x: x > 5],
     alpha=[1, .7],
     colorbar=True,
    )
```

```python
plot(data=[arr, [arr, arr]],
     title=['t1', 't2'],
     cmap=['Greys', ['Greys', 'Blues']],
     mask_values=[None, [None, lambda x: x > 5]],
     alpha=[1, [1, .7]],
     colorbar=True,
    )
```

```python
plot(data=[arr, [arr, arr < 5]],
     title=['t1', 't2'],
     cmap=['Greys', ['Greys', 'firebrick']],
     alpha=[1, [1, .7]],
     colorbar=True,
    )
```

```python
diverse_data = [[np.random.rand(*np.random.randint(2, 10, 2))] for _ in range(np.random.randint(5, 10))]
plot(diverse_data)
```
